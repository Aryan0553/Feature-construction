```python
import numpy as np 
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns 
```

```python
df=pd.read_csv("C:/Program Files (x86)/New excel/tested.csv")[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#seprate
X = df.iloc[:,0:4]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
#cross validate 20 times 
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

```python
#feature construction apply
X['Family_Size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0 
    elif num >1 and num <=4:
        #small family 
        return 1 
    else:
        #large family
        return 2
        
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_Size'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_Size'], inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```
